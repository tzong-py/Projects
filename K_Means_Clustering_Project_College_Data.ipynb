# **K Means Cluster:**

```python
from sklearn.cluster import KMeans
```

```python
#Create 2 clusters since we want to divide them into Private or Public:
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(coldata.drop('Private', axis=1))
```

```python
kmeans.cluster_centers_
```

**Evaluation**

In real world we cannot make this evaluation since we don't have the label like provided in this exercise

```python
def converter(cluster):
  if cluster =='Yes':
    return 1
  else:
    return 0
```

```python
coldata['cluster'] = coldata['Private'].apply(converter)
```

```python
coldata.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(coldata['cluster'],kmeans.labels_))
print(classification_report(coldata['cluster'],kmeans.labels_))
```
